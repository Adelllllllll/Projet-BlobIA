```python
import os
import sys
import pandas as pd

# Si tu lances le notebook depuis notebooks/
os.chdir("..")

# Charger les données
monuments = pd.read_csv("data/BDD Monument Gare Affluence.csv")
stations = pd.read_csv("data/emplacement-des-gares-idf.csv")

# Récupère la liste des stations avec leur latitude
stations_coords = stations[['nom_so_gar', 'Geo Point']].drop_duplicates()

# Liste triée des stations du sud au nord
stations_sorted_sud = stations_coords.sort_values(by="Geo Point", ascending=True)['nom_so_gar'].tolist()
stations_sorted_nord = stations_coords.sort_values(by="Geo Point", ascending=False)['nom_so_gar'].tolist()

# Fonction qui trouve la première station extrême présente dans les monuments
def find_station_in_monuments(station_list, monuments_df):
    for station in station_list:
        if station in monuments_df['Station la plus proche'].values:
            return station
    return None

# Trouver une station très au sud ET une station très au nord qui existent comme "Station la plus proche" d'au moins un monument
station_depart = find_station_in_monuments(stations_sorted_sud, monuments)
station_arrivee = find_station_in_monuments(stations_sorted_nord, monuments)

# Trouver un monument associé à chacune de ces stations
monument_depart = monuments[monuments['Station la plus proche'] == station_depart]['Monument'].iloc[0]
monument_arrivee = monuments[monuments['Station la plus proche'] == station_arrivee]['Monument'].iloc[0]

print(f"Couple exemple :")
print(f"- Départ : {station_depart} (monument : {monument_depart})")
print(f"- Arrivée : {station_arrivee} (monument : {monument_arrivee})")
```
