```python
import pandas as pd

# 1. Chargement de tous les fichiers
affluence = pd.read_excel("data/affluence_stations_paris.xlsx")
monuments = pd.read_csv("data/BDD Monument Gare Affluence.csv")
gares = pd.read_csv("data/emplacement-des-gares-idf.csv")
stations = pd.read_csv("data/Stations IDF.csv")
```

```python
# 2. Afficher les dimensions et les noms de colonnes de chaque dataset
print("Affluence - shape:", affluence.shape)
print("Colonnes:", affluence.columns)
print("\nMonuments - shape:", monuments.shape)
print("Colonnes:", monuments.columns)
print("\nGares - shape:", gares.shape)
print("Colonnes:", gares.columns)
print("\nStations - shape:", stations.shape)
print("Colonnes:", stations.columns)
```

```python
# 3. Aperçu des premières lignes de chaque dataset
print("\n--- Aperçu Affluence ---")
display(affluence.head())
print("\n--- Aperçu Monuments ---")
display(monuments.head())
print("\n--- Aperçu Gares ---")
display(gares.head())
print("\n--- Aperçu Stations ---")
display(stations.head())
```

```python
# 4. Statistiques descriptives basiques
print("\n--- Statistiques Affluence ---")
display(affluence.describe(include='all'))
print("\n--- Statistiques Monuments ---")
display(monuments.describe(include='all'))
print("\n--- Statistiques Gares ---")
display(gares.describe(include='all'))
print("\n--- Statistiques Stations ---")
display(stations.describe(include='all'))
```

```python
# 5. Vérification des valeurs manquantes
print("\n--- Valeurs manquantes Affluence ---")
print(affluence.isnull().sum())
print("\n--- Valeurs manquantes Monuments ---")
print(monuments.isnull().sum())
print("\n--- Valeurs manquantes Gares ---")
print(gares.isnull().sum())
print("\n--- Valeurs manquantes Stations ---")
print(stations.isnull().sum())
```

```python
import pandas as pd

# Charge le CSV (modifie le chemin selon ton projet)
df = pd.read_csv("data/Stations_IDF_aligned.csv")

# S'assure que tout est bien str
df["station_key"] = df["station_key"].astype(str)
df["ligne"] = df["ligne"].astype(str)

# Calcule le nombre de lignes uniques pour chaque station_key
corresp = df.groupby("station_key")["ligne"].nunique().reset_index()
corresp.columns = ["station_key", "nb_lignes"]

# Ne garde que les correspondances (apparaît sur >1 ligne)
corresp = corresp[corresp["nb_lignes"] > 1]

# Trie du plus grand au plus petit
corresp = corresp.sort_values("nb_lignes", ascending=False)

print("Liste des stations de correspondance (multi-lignes) :")
print(corresp)

# Optionnel : exporte la liste pour traitement manuel/futur
corresp.to_csv("data/stations_correspondance.csv", index=False)
print("\nFichier créé : data/stations_correspondance.csv")
```
